```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager


# -----------------------
# Setup Chrome Driver
# -----------------------
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = "https://collegedunia.com/btech-colleges"
driver.get(url)

wait = WebDriverWait(driver, 30)

# Wait until first rows load
wait.until(EC.presence_of_element_located(
    (By.CSS_SELECTOR, "table.listing-table tbody tr.table-row")
))

# -----------------------
# Scroll Until All Rows Load
# -----------------------
last_count = 0

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

    rows = driver.find_elements(By.CSS_SELECTOR, "table.listing-table tbody tr.table-row")
    current_count = len(rows)

    print("Rows currently loaded:", current_count)

    if current_count == last_count:
        break

    last_count = current_count

print("Scrolling complete. Raw rows found:", last_count)

# -----------------------
# Extract & Remove Duplicates
# -----------------------
unique_colleges = {}

rows = driver.find_elements(By.CSS_SELECTOR, "table.listing-table tbody tr.table-row")

for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")

    if len(cols) < 6:
        continue

    rank = cols[0].text.strip()
    college_info = cols[1].text.strip()
    fees = cols[2].text.strip()
    placement = cols[3].text.strip()
    reviews = cols[4].text.strip()
    ranking = cols[5].text.strip()

    # Extract college name (first line only)
    college_name = college_info.split("\n")[0]

    # Remove duplicates using college name
    if college_name not in unique_colleges:
        unique_colleges[college_name] = {
            "Rank": rank,
            "College Name": college_name,
            "Full College Info": college_info,
            "Course Fees": fees,
            "Placement": placement,
            "User Reviews": reviews,
            "Ranking": ranking
        }

driver.quit()

# -----------------------
# Save Data
# -----------------------
df = pd.DataFrame(unique_colleges.values())
df.to_csv("new_collegedunia_btech_all_clean.csv", index=False)

print("✅ Final Unique Colleges Scraped:", len(df))
print(df.head())
```
